```python
#繪製(黃)色x標記
plt.plot([9,9.2,9.6,7.5,6.7,7],[9.4,9.2,9.2,9.2,7.1,7.4],'yx')
plt.plot([9,9.2,9.6,7.5,6.7,7],[9.4,9.2,9.2,9.2,7.1,7.4],'yx')
#繪製(綠)色x標記
plt.plot([7.2,7.3,7.2,7.3,7.2,7.3,7.3],[10.3,10.5,9.2,10.2,9.7,10.1,10.1],'gx')
plt.plot([6.5,9.0],[7.8,12.5],'b--') # 繪製黑色虛線
plt.ylabel('H cm') # 設定顯示Y文字
plt.xlabel('W cm') # 設定顯示X文字

#繪製圖表
plt.legend(('Oranage','Lemons'),loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt # 繪圖函示庫

plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'gx') #在x=1,y=0四個位置繪製(g)色 標記
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'r--') #繪製(r)紅色 標記也就是綠(趨勢線)
plt.axis([0,5,0,1]) #圖表大小範圍寬度由0到5，高度由0到1
plt.ylabel('Y') #設定顯示Y 文字
plt.xlabel('X') #設定顯示X 文字
plt.legend(('price','passenger'),loc='upper right')
plt.show() #繪製圖表
```

```python
import matplotlib.pyplot as plt # 繪圖函示庫
import numpy as np
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'gx') # 綠色，x的點
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'r--') # 紅色，--虛線

#取亂數
X = 1+np.arange(30)/10
delta = np.random.uniform(low=-0.1,high=0.1,size=(30,))
#帶入公式
Y=0.3*X- 0.3 + delta
plt.plot(X,Y, 'bo') #藍色的圓點
plt.xlabel('X') #設定顯示X文字
plt.ylabel('Y') #設定顯示Y文字
plt.show() #繪圖
```
